```python
import pandas as pd
from sklearn.datasets import load_iris, load_wine
iris_data = load_iris()
wine_data = load_wine()
```

```python
def categorize_columns(data, column_names):
    numeric_cols = []
    categorical_cols = []
    
   
    for col in range(len(data[0])):
        if isinstance(data[0][col], (int, float)):
            numeric_cols.append(column_names[col])
        else:
            categorical_cols.append(column_names[col])

    print("Numeric Columns:", numeric_cols)
    print("Categorical Columns:", categorical_cols)


iris_data = [
    [5.1, 3.5, 1.4, 0.2, 'setosa'],
    [4.9, 3.0, 1.4, 0.2, 'setosa'],
    [6.2, 3.4, 5.4, 2.3, 'virginica'],
]
iris_column_names = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']

wine_data = [
    [13.2, 1.78, 2.14, 11.2, 'class_0'],
    [12.3, 1.63, 2.02, 9.8, 'class_1'],
    [11.6, 1.99, 1.95, 10.3, 'class_2'],
]
wine_column_names = ['alcohol', 'malic_acid', 'ash', 'alcalinity', 'class']


categorize_columns(iris_data, iris_column_names)
categorize_columns(wine_data, wine_column_names)
```

## In NumPy:
Calculate the correlation matrix for all numeric columns.
Identify the pair of columns with the highest correlation coefficient.
Print the names of these columns along with their correlation coefficient.
Include a section in your report where you discuss your comprehension of this task and its relevance in the field of data pre-processing and analysis using pandas and NumPy libraries. Make sure to run the code and attach a screenshot of your machine in the appendix of your report.

```python
def correlation_matrix(df):
    correlation_matrix = df.corr()

    max_corr_idx = np.unravel_index(np.argmax(np.abs(correlation_matrix.values)), correlation_matrix.shape)

    name_col1 = correlation_matrix.columns[max_corr_idx[0]]
    name_col2 = correlation_matrix.columns[max_corr_idx[1]]
    highest_corr = correlation_matrix.iloc[max_corr_idx]

    print(name_col1, name_col2, highest_corr)
    
    return name_col1, name_col2, highest_corr
```

```python
name_col1, name_col2, highest_corr = correlation_matrix(iris_df)
```

## Q5 Conduct an exploratory data analysis (EDA) on the Iris dataset from scikit learn using Python. The objective is to provide a comprehensive overview of the dataset, including an examination of its features and target variables, as well as understanding the data structure—specifically its shape, data types, and any missing values. The analysis should include descriptive statistics for each feature, along with visualizations such as histograms and boxplots to explore feature distributions and relationships among the species. Additionally, a correlation matrix should be generated to identify significant correlations between features. Finally, the report should include interpretations and conclusions drawn from the analyses and visualizations, effectively summarizing insights gained from the EDA process. Make sure to run the code and attach a screenshot of your machine in the appendix of your report.

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris

iris = load_iris()
df = pd.DataFrame(data=iris.data, columns=iris.feature_names)
df['species'] = iris.target
df['species'] = df['species'].map({i: species for i, species in enumerate(iris.target_names)})


print("Descriptive Statistics:", df.describe())
```

```python
df.hist(figsize=(12, 8))
plt.show()
```

```python
plt.figure(figsize=(10, 8))
sns.boxplot(data=df, orient="h", palette="Set2")
plt.title("boxplots")
plt.show()
```

```python
corr_matrix = df.drop(columns=['species']).corr()
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(df, hue="species", palette="Set1", markers=["o", "s", "D"])
plt.suptitle("Relationships by Species")
plt.show()
```
